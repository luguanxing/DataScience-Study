## DBSCAN聚类DBSCANclustering

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt  
import numpy as np      

# 生成数据 generation data
X, y = make_blobs(n_samples=20, centers=1, random_state=0, cluster_std=1.00)  # 生成20个样本，2个中心。Generate 20 samples with 2 centers
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')                      # 绘制散点图。Plot scatter plot
plt.title('Original Data')                                                     # 图标题。Plot title
plt.xlabel('Feature 1')                                                        # x轴标签。x-axis label
plt.ylabel('Feature 2')                                                        # y轴标签。y-axis label
plt.show()                                                                      # 显示图形。Show plot
```

```python
# 使用DBSCAN进行聚类 Use DBSCAN for clustering
dbscan = DBSCAN(eps=0.5, min_samples=3)  # 设置eps和min_samples参数。Set eps and min_samples
y_pred = dbscan.fit_predict(X)  # 预测聚类标签。Predict cluster labels

# 绘制散点图，使用DBSCAN的聚类结果 Plot scatter plot with DBSCAN clustering result
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap='autumn')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# 设置颜色 set colors based on cluster labels
colors = ['blue', 'green']

# 生成随机数据 generation random data
rX, ry = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=1.00)  # 生成100个样本，2个中心。Generate 20 samples with 2 centers

# 绘制所有点的散点图，填充颜色根据 `y` 自动使用 `cmap` 映射
# Plot all points in a scatter plot, with fill colors mapped according to 'y' using 'cmap'
scatter = plt.scatter(rX[:, 0], rX[:, 1], c=ry, s=50, cmap='autumn', linewidth=2)

# 手动设置每个点的边框颜色（需要逐个点绘制）
# manually set the edge color for each point
for i in range(len(rX)):
    c = dbscan.fit_predict(X)
    print(c)
    plt.scatter(rX[i, 0], rX[i, 1], edgecolor=colors[c], s=50, facecolors='none', linewidth=2)
```
