```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# 读取数据
# Read data
votes = pd.read_csv("./114_congress.csv")

# 打印数据的维度（行数和列数）
# Print the shape of the dataset (number of rows and columns)
print(votes.shape)

# 显示数据内容
# Display the dataset
votes
```

```python
# 探索数据
# Explore the data
print(votes["party"].value_counts())

votes.drop(['name', 'party', 'state'], axis=1).mean().to_frame().T
```

```python
# 使用K-means算法创建2个聚类
# Use K-means to create 2 clusters
from sklearn.cluster import KMeans

# 计算分类
# Compute the classification
numeric_votes = votes.drop(['name', 'party', 'state'], axis=1)
kmeans2 = KMeans(n_clusters=2, random_state=42)
votes['class2'] = kmeans2.fit_predict(numeric_votes)

# 计算到聚类中心的距离
# Calculate distances to centroids
distances = kmeans2.transform(numeric_votes)
votes['class2_x'] = distances[:, 0]
votes['class2_y'] = distances[:, 1]

votes.head()
```

```python
# 检查聚类与政党之间的关系
# Check relation between clusters and parties
labels = kmeans2.labels_
pd.crosstab(labels, votes["party"])
```

```python
# 找出实际为民主党但被分到共和党类别class2=1的议员
# Find congressmen who are actually Democrats but were assigned to Republican class (class2=1)
apostate = votes[(votes['party'] == 'D') & (votes['class2'] == 1)]

# 输出这些议员的名字
# Output the names of these congressmen
apostate
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 绘制基础散点图，设置 class2=0 为蓝色，class2=1 为红色
# 1. Plot a basic scatter plot, setting class2=0 in blue and class2=1 in red
plt.figure(figsize=(10, 6))
sns.scatterplot(x='class2_x', y='class2_y', hue='class2', data=votes, 
                palette={0: 'blue', 1: 'red'}, s=100)

# 2. 给每个点标注其对应的党派（D, R, I）
# 2. Label each point with its corresponding party (D, R, I)
for idx, row in votes.iterrows():
    plt.text(row['class2_x'], row['class2_y'], row['party'], 
             color='pink', fontsize=10, ha='center', va='center', weight='bold')

# 3. 特别标记“卧底”议员
# 3. Specially mark the "apostate" congressmen
for idx, row in apostate.iterrows():
    plt.text(row['class2_x'], row['class2_y']+0.1, row['name'], 
             color='black', fontsize=12, weight='bold', ha='center', va='center')
    plt.scatter(row['class2_x'], row['class2_y'], s=200, edgecolor='black', label=row['name'])

# 4. 添加标题和标签
# 4. Add title and labels
plt.title('2D Plot of Congressmen Distances to Cluster Centroids (with Party Labels)')
plt.xlabel('Distance to Centroid 1 (class2_x)')
plt.ylabel('Distance to Centroid 2 (class2_y)')

# 5. 显示图形
# 5. Show the plot
plt.legend(loc='upper right')
plt.show()
```
