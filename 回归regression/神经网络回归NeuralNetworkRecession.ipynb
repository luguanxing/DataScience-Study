```python
# data generation
import numpy as np
import matplotlib.pyplot as plt

# 生成随机的x值 (Generate random x values)
x_values = np.random.uniform(-10, 10, 20)

# 计算直线上的y值 y = 3x + 5 (Calculate y values on the line y=3x+5)
y_values = 3 * x_values + 5

# 添加随机噪声，使点分布在直线附近 (Add random noise to make the points distributed around the line)
noise = np.random.normal(0, 3, 20)  # 均值为0，标准差为3的噪声 (Noise with mean 0 and standard deviation 3)
y_random = y_values + noise

# 绘制直线 y = 3x + 5 (Plot the line y = 3x + 5)
x_line = np.linspace(-10, 10, 100)
y_line = 3 * x_line + 5
plt.plot(x_line, y_line, color='red', label='y = 3x + 5')

# 绘制随机点 (Plot the random points)
plt.scatter(x_values, y_random, color='blue', label='Random points near the line')

# 添加图例和标签 (Add legend and labels)
plt.legend()
plt.title('Random Points Near the Line y=3x+5')  # 图形标题 (Plot title)
plt.xlabel('x')  # x轴标签 (x-axis label)
plt.ylabel('y')  # y轴标签 (y-axis label)

# 显示图形 (Show the plot)
plt.grid(True)
plt.show()
```

```python
# 导入必要的库 (Import necessary libraries)
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2

# 构建神经网络模型 (Build the neural network model)
model = Sequential()
model.add(Dense(64, input_dim=1, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dense(1))

# 编译模型 (Compile the model)
model.compile(optimizer='adam', loss='mean_squared_error')

# 将x_values转换为二维数组，适应神经网络输入要求 (Reshape x_values to 2D array for neural network input)
x_train = x_values.reshape(-1, 1)
y_train = y_random

# 训练模型 (Train the model)
model.fit(x_train, y_train, epochs=500, verbose=0)  # 训练500轮 (Train for 500 epochs)

# 使用训练好的模型进行预测 (Use the trained model for prediction)
x_line = np.linspace(-10, 10, 100).reshape(-1, 1)
y_pred = model.predict(x_line)

print(model)
```

```python
# 使用训练好的模型进行预测 (Use the trained model for prediction)
x_line = np.linspace(-10, 10, 100).reshape(-1, 1)
y_pred = model.predict(x_train)
y_line_pred = model.predict(x_line)

# 绘制原始的直线和随机点 (Plot the original line and the random points)
plt.plot(x_line, 3 * x_line + 5, color='red', label='y = 3x + 5 (True Line)')
plt.scatter(x_values, y_random, color='red', label='Random points near the line')

# 绘制神经网络的预测结果 (Plot the neural network predictions)
plt.plot(x_line, y_line_pred, color='green', label='NN Prediction (Neural Network)')
plt.scatter(x_values, y_pred, color='green', marker='x', linewidths=3.0, label='NN Predictions (Neural Network)', s=100)

# 添加图例和标签 (Add legend and labels)
plt.legend()
plt.title('Neural Network Regression Prediction')  # 图形标题 (Plot title)
plt.xlabel('x')  # x轴标签 (x-axis label)
plt.ylabel('y')  # y轴标签 (y-axis label)

# 显示图形 (Show the plot)
plt.grid(True)
plt.show()
```
