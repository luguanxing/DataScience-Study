## K均值聚类K-means

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt  
import numpy as np      

# 生成数据 generation data
X, y = make_blobs(n_samples=20, centers=2, random_state=0, cluster_std=1.00)  # 生成20个样本，2个中心。Generate 20 samples with 2 centers
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')                      # 绘制散点图。Plot scatter plot
plt.title('Original Data')                                                     # 图标题。Plot title
plt.xlabel('Feature 1')                                                        # x轴标签。x-axis label
plt.ylabel('Feature 2')                                                        # y轴标签。y-axis label
plt.show()                                                                      # 显示图形。Show plot
```

```python
# 应用KMeans聚类 / Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=0)              # 初始化KMeans，聚类数为2。Initialize KMeans with 2 clusters
kmeans.fit(X)                                              # 训练模型。Fit the model
centroids = kmeans.cluster_centers_                        # 获取质心。Get centroids
labels = kmeans.labels_                                    # 获取聚类标签。Get cluster labels

# 定义函数根据点预测所属聚类 / Define a function to predict the cluster of a point based on KMeans
def predict_cluster(point, kmeans_model):
    point = np.array(point).reshape(1, -1)                 # 将点转换为二维数组。Convert the point to a 2D array
    cluster_label = kmeans_model.predict(point)[0]         # 使用KMeans预测聚类标签。Predict the cluster label using KMeans
    return cluster_label                                    # 返回聚类标签。Return the cluster label

print(predict_cluster(centroids[0], kmeans))
print(predict_cluster(centroids[1], kmeans))
```

```python
# 设置边框颜色 set edge colors based on cluster labels
colors = ['blue', 'green']

# 绘制质心 Plot centroids
for _, centroid in enumerate(centroids):
    plt.scatter(centroid[0], centroid[1], marker='^', s=200, c=colors[predict_cluster(centroid, kmeans)], edgecolor=colors[predict_cluster(centroid, kmeans)])

# 绘制所有点的散点图，填充颜色根据 `y` 自动使用 `cmap` 映射
# Plot all points in a scatter plot, with fill colors mapped according to 'y' using 'cmap'
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn', linewidth=2)

# 手动设置每个点的边框颜色（需要逐个点绘制）
# manually set the edge color for each point
for i in range(len(X)):
    c = predict_cluster(X[i], kmeans)
    plt.scatter(X[i, 0], X[i, 1], edgecolor=colors[c], s=50, facecolors='none', linewidth=2)

# 绘制聚类结果 plot clustering results
plt.title('KMeans Clustering Result')                                                            # 图标题。Plot title
plt.xlabel('Feature 1')                                                                          # x轴标签。x-axis label
plt.ylabel('Feature 2')                                                                          # y轴标签。y-axis label
plt.show()                                                                                        # 显示图形。Show plot
```
