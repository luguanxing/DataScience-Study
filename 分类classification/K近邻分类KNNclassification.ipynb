```python
# data generation
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from sklearn.svm import SVC

X, y = make_circles(100, factor=.1, noise=.1)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 创建KNN分类器 create knn classificator
knn = KNeighborsClassifier(n_neighbors=3)

# 训练模型 train model
knn.fit(X, y)
```

```python
from sklearn.datasets import make_blobs
import numpy as np

# 生成随机数据 generation random data
n = 30  # 随机点数量
ax = np.random.uniform(-1, 1, n)
ay = np.random.uniform(-1, 1, n)
rX = np.column_stack((ax, ay))

# 进行预测
y_pred = knn.predict(rX)

# 设置颜色 set colors based on cluster labels
colors = ['orange', 'green']

# 手动设置每个点的边框颜色（需要逐个点绘制）
# manually set the edge color for each point
for i in range(len(rX)):
    c = y_pred[i]
    plt.scatter(rX[i, 0], rX[i, 1], color=colors[c], s=50, marker='x', linewidth=2)

# 画原始数据 draw original data
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.title('Original Data')                                                     # 图标题。Plot title
plt.xlabel('Feature 1')                                                        # x轴标签。x-axis label
plt.ylabel('Feature 2')                                                        # y轴标签。y-axis label
plt.show()                                                                      # 显示图形。Show plot
```
